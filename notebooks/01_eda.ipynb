```python
import numpy as np

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)

# Helper utilities to keep EDA consistent and safe on large tables
TIME_COL_HINTS = (
    "date",
    "time",
    "timestamp",
    "created",
    "updated",
    "start",
    "end",
    "occurred",
    "measured",
    "admit",
    "discharge",
)


def detect_time_cols(df: pd.DataFrame) -> list[str]:
    cols = []
    for c in df.columns:
        lc = str(c).lower()
        if any(h in lc for h in TIME_COL_HINTS):
            cols.append(c)
    return cols


def summarize_table(df: pd.DataFrame, sample_rows: int = 200_000) -> dict:
    n_rows, n_cols = df.shape
    mem_mb = df.memory_usage(deep=True).sum() / (1024 ** 2)
    time_cols = detect_time_cols(df)
    sample = df
    if n_rows > sample_rows:
        sample = df.sample(sample_rows, random_state=42)
    missing_mean = float(sample.isna().mean().mean())
    return {
        "rows": n_rows,
        "cols": n_cols,
        "mem_mb": round(mem_mb, 2),
        "missing_mean": round(missing_mean, 4),
        "time_cols": ", ".join(time_cols) if time_cols else "",
        "has_resident_id": "resident_id" in df.columns,
        "has_facility_id": "facility_id" in df.columns,
    }


summary_rows = []
for name, df in dfs.items():
    row = {"table": name}
    row.update(summarize_table(df))
    summary_rows.append(row)

summary_df = (
    pd.DataFrame(summary_rows)
    .sort_values(["rows", "cols"], ascending=False)
    .reset_index(drop=True)
)

summary_df
```

```python
# Timestamp outlier scan (sampling for large tables)
outlier_rows = []
for name, df in dfs.items():
    time_cols = detect_time_cols(df)
    if not time_cols:
        continue
    n_rows = len(df)
    sample = df if n_rows <= 200_000 else df.sample(200_000, random_state=42)
    for col in time_cols:
        series = pd.to_datetime(sample[col], errors="coerce")
        if series.notna().sum() == 0:
            continue
        min_val = series.min()
        max_val = series.max()
        # Simple heuristic: flag if outside plausible range
        if min_val.year < 2000 or max_val.year > 2026:
            outlier_rows.append(
                {
                    "table": name,
                    "column": col,
                    "min": min_val,
                    "max": max_val,
                    "n_nonnull": int(series.notna().sum()),
                }
            )

outlier_df = pd.DataFrame(outlier_rows).sort_values(["table", "column"])
outlier_df
```

```python
# Date range out-of-range counts by table (no parsing fixes)
min_year = 2000
max_year = 2026

exclude_cols = {"date_of_birth", "dob", "birth_date", "birthdate"}

rows = []
for name, df in dfs.items():
    time_cols = detect_time_cols(df)
    if not time_cols:
        continue
    for col in time_cols:
        if str(col).lower() in exclude_cols:
            continue
        ts = pd.to_datetime(df[col], errors="coerce")
        out_mask = ((ts.dt.year < min_year) | (ts.dt.year > max_year)) & ts.notna()
        rows.append(
            {
                "table": name,
                "column": col,
                "out_of_range": int(out_mask.sum()),
                "total_nonnull": int(ts.notna().sum()),
                "min": ts.min(),
                "max": ts.max(),
            }
        )

out_range_df = (
    pd.DataFrame(rows)
    .sort_values(["out_of_range", "total_nonnull"], ascending=False)
    .reset_index(drop=True)
)

out_range_df
```

```python
# Filter out rows with date-like columns outside 2000-01 to 2026-12 (keep raw dfs)
min_year = 2000
max_year = 2026

exclude_cols = {"date_of_birth", "dob", "birth_date", "birthdate"}

dfs_filtered = {}
filter_summary = []

for name, df in dfs.items():
    keep = pd.Series(True, index=df.index)
    for col in detect_time_cols(df):
        if str(col).lower() in exclude_cols:
            continue
        ts = pd.to_datetime(df[col], errors="coerce")
        in_range = (ts.dt.year >= min_year) & (ts.dt.year <= max_year)
        keep &= in_range | ts.isna()

    filtered = df.loc[keep].copy()
    dfs_filtered[name] = filtered

    filter_summary.append(
        {
            "table": name,
            "rows_before": len(df),
            "rows_after": len(filtered),
            "dropped": int(len(df) - len(filtered)),
        }
    )

filter_summary_df = pd.DataFrame(filter_summary).sort_values("dropped", ascending=False)
filter_summary_df
```

```python
# Find remaining incident categories and estimate losses using average costs

avg_costs = {
    "Fall": 3500,
    "Medication Error": 5000,
    "Wound": 4000,
    "Return-to-hospital": 20000,
    "Elopement": 2500,
    "Altercation": 2500,
}

source_dfs = dfs_filtered if "dfs_filtered" in globals() else dfs

counts = {}

# From incidents table
if "incidents" in source_dfs:
    inc = source_dfs["incidents"].copy()
    if "incident_type" in inc.columns:
        type_counts = inc["incident_type"].value_counts(dropna=False)
        for key in ["Fall", "Medication Error", "Wound", "Elopement", "Altercation"]:
            counts[key] = int(type_counts.get(key, 0))

# Return-to-hospital: 30-day readmission from hospital_admissions
rth_source = ""
rth_method = ""
rth_detail = ""
rth_count = 0

if "hospital_admissions" in source_dfs:
    ha = source_dfs["hospital_admissions"].copy()
    rth_source = "hospital_admissions"

    from typing import Optional

    def pick_first_col(df: pd.DataFrame, candidates: list[str]) -> Optional[str]:
        for c in candidates:
            if c in df.columns:
                return c
        return None

    admit_col = pick_first_col(
        ha,
        [
            "effective_date",
            "admission_date",
            "admit_date",
            "admitted_at",
            "admit_at",
            "start_at",
            "start_date",
        ],
    )
    discharge_col = pick_first_col(
        ha,
        [
            "ineffective_date",
            "hospital_stay_to",
            "discharge_date",
            "discharged_at",
            "discharge_at",
            "end_at",
            "end_date",
        ],
    )
    status_col = pick_first_col(
        ha,
        [
            "admission_status",
            "stay_purpose",
            "admission_type",
            "transfer_reason",
            "reason",
        ],
    )

    if admit_col and discharge_col and "resident_id" in ha.columns:
        ha["admit_ts"] = pd.to_datetime(ha[admit_col], errors="coerce")
        ha["discharge_ts"] = pd.to_datetime(ha[discharge_col], errors="coerce")

        planned_mask = pd.Series(False, index=ha.index)
        if status_col:
            status = ha[status_col].astype(str).str.lower()
            planned_mask = status.str.contains("planned|elective|scheduled", na=False)

        ha_sorted = ha.sort_values(["resident_id", "admit_ts"])
        ha_sorted["prev_discharge_ts"] = ha_sorted.groupby("resident_id")["discharge_ts"].shift(1)

        readmit_mask = (
            ha_sorted["admit_ts"].notna()
            & ha_sorted["prev_discharge_ts"].notna()
            & (ha_sorted["admit_ts"] >= ha_sorted["prev_discharge_ts"])
            & (
                ha_sorted["admit_ts"]
                <= ha_sorted["prev_discharge_ts"] + pd.Timedelta(days=30)
            )
        )
        if status_col:
            readmit_mask &= ~planned_mask.loc[ha_sorted.index]

        rth_count = int(readmit_mask.sum())
        rth_method = "30-day readmission from discharge to next admission"
        if status_col:
            rth_method += f"; exclude planned via {status_col}"
        else:
            rth_method += "; no planned flag"
        rth_detail = f"admit={admit_col}, discharge={discharge_col}"
    else:
        rth_method = "hospital_admissions missing required columns"
else:
    rth_method = "hospital_admissions not available"

counts["Return-to-hospital"] = int(rth_count)
```

```python
# Build loss estimate table
rows = []
for k, cost in avg_costs.items():
    n = counts.get(k, 0)
    rows.append({"incident_type": k, "count": n, "avg_cost": cost, "est_loss": n * cost})

loss_df = pd.DataFrame(rows)
loss_df.loc["TOTAL"] = {
    "incident_type": "TOTAL",
    "count": int(loss_df["count"].sum()),
    "avg_cost": "",
    "est_loss": int(loss_df["est_loss"].sum()),
}

print("RTH method:", rth_source, "-", rth_method)
if rth_detail:
    print("RTH columns:", rth_detail)

display(loss_df.sort_values("est_loss", ascending=False))
```

```python
# Build a per-resident timeline for hospital events only (admissions + transfers)
base_dfs = dfs_filtered if "dfs_filtered" in globals() else dfs
source_dfs = {
    k: v for k, v in base_dfs.items() if k in {"hospital_admissions", "hospital_transfers"}
}

def build_resident_timeline(resident_id, source_dfs):
    rows = []
    for name, df in source_dfs.items():
        if "resident_id" not in df.columns:
            continue
        subset = df[df["resident_id"] == resident_id].copy()
        if subset.empty:
            continue
        time_cols = detect_time_cols(subset)
        if not time_cols:
            continue
        time_df = subset[time_cols].apply(pd.to_datetime, errors="coerce")
        subset["global_time"] = time_df.min(axis=1)
        subset["source_table"] = name
        rows.append(subset)

    if not rows:
        return pd.DataFrame()

    timeline = pd.concat(rows, ignore_index=True)
    timeline = timeline.sort_values(["global_time", "source_table"], na_position="last")
    return timeline

# Set a resident_id manually (example: resident_id = 12345)
resident_id = "072691ff-60ba-5d4c-a8b4-ba582a99d803"
if resident_id is None:
    raise ValueError("Set resident_id to a valid value before running.")

timeline = build_resident_timeline(resident_id, source_dfs)
display(timeline.head(200))

# # Export full timeline to CSV
# if not timeline.empty:
#     output_path = Path("../outputs") / f"timeline_{resident_id}.csv"
#     timeline.to_csv(output_path, index=False)
#     print("Saved:", output_path)
# else:
#     print("No timeline rows to export.")
```

```python
# Hospital event ordering metrics using effective/ineffective dates
base_dfs = dfs_filtered if "dfs_filtered" in globals() else dfs
ha = base_dfs.get("hospital_admissions", pd.DataFrame()).copy()
ht = base_dfs.get("hospital_transfers", pd.DataFrame()).copy()

events = []
if not ha.empty:
    ha_events = ha.copy()
    ha_events["event_type"] = "admission"
    ha_events["effective_date"] = pd.to_datetime(ha_events.get("effective_date"), errors="coerce")
    ha_events["ineffective_date"] = pd.to_datetime(ha_events.get("ineffective_date"), errors="coerce")
    events.append(ha_events)
if not ht.empty:
    ht_events = ht.copy()
    ht_events["event_type"] = "transfer"
    ht_events["effective_date"] = pd.to_datetime(ht_events.get("effective_date"), errors="coerce")
    ht_events["ineffective_date"] = pd.to_datetime(ht_events.get("ineffective_date"), errors="coerce")
    events.append(ht_events)

if not events:
    raise ValueError("hospital_admissions and hospital_transfers are empty or missing.")

events_df = pd.concat(events, ignore_index=True)
events_df = events_df[events_df["resident_id"].notna()].copy()
events_df = events_df[events_df["effective_date"].notna()].copy()

events_df = events_df.sort_values(["resident_id", "effective_date"])
events_df["next_event_type"] = events_df.groupby("resident_id")["event_type"].shift(-1)
events_df["next_effective_date"] = events_df.groupby("resident_id")["effective_date"].shift(-1)

# Adjacent pattern counts
adj_mask = events_df["next_event_type"].notna()
adj_pairs = (
    events_df.loc[adj_mask, "event_type"]
    + "->"
    + events_df.loc[adj_mask, "next_event_type"]
)
adj_pair_counts = adj_pairs.value_counts().to_frame("count")
total_adj = int(adj_pair_counts["count"].sum())
adj_pair_counts["share"] = adj_pair_counts["count"] / total_adj if total_adj else 0.0
display(adj_pair_counts)

# Gap metrics for transfer->admission and admission->transfer
events_df["gap_days"] = (
    events_df["next_effective_date"] - events_df["ineffective_date"]
).dt.days
gap_ta = events_df[(events_df["event_type"] == "transfer") & (events_df["next_event_type"] == "admission")].copy()
gap_at = events_df[(events_df["event_type"] == "admission") & (events_df["next_event_type"] == "transfer")].copy()

def gap_summary(df: pd.DataFrame, label: str) -> dict:
    valid = df[df["gap_days"].notna()]
    return {
        "pattern": label,
        "pairs": int(len(df)),
        "gaps_with_dates": int(len(valid)),
        "median_gap_days": float(valid["gap_days"].median()) if not valid.empty else np.nan,
        "p10_gap_days": float(valid["gap_days"].quantile(0.1)) if not valid.empty else np.nan,
        "p90_gap_days": float(valid["gap_days"].quantile(0.9)) if not valid.empty else np.nan,
        "negative_gap_count": int((valid["gap_days"] < 0).sum()) if not valid.empty else 0,
    }

gap_summary_df = pd.DataFrame(
    [
        gap_summary(gap_ta, "transfer->admission"),
        gap_summary(gap_at, "admission->transfer"),
    ]
).set_index("pattern")
display(gap_summary_df)

# Resident-level dominant pattern
resident_pairs = events_df.loc[adj_mask, ["resident_id", "event_type", "next_event_type"]].copy()
resident_pairs["pair"] = resident_pairs["event_type"] + "->" + resident_pairs["next_event_type"]
resident_counts = (
    resident_pairs.groupby(["resident_id", "pair"]).size().unstack(fill_value=0)
)
resident_counts["total_pairs"] = resident_counts.sum(axis=1)
pair_cols = [c for c in resident_counts.columns if c != "total_pairs"]
resident_counts["top_pair"] = resident_counts[pair_cols].idxmax(axis=1)
resident_counts["top_pair_share"] = (
    resident_counts[pair_cols].max(axis=1) / resident_counts["total_pairs"]
).fillna(0)

top_pair_dist = resident_counts["top_pair"].value_counts().to_frame("residents")
top_pair_dist["share"] = top_pair_dist["residents"] / top_pair_dist["residents"].sum()
display(top_pair_dist)

# Transfer duration hypothesis check
first_event = events_df.groupby("resident_id")["event_type"].first().rename("first_event_type")
events_df = events_df.merge(first_event, on="resident_id", how="left")
events_df["transfer_duration_days"] = (
    events_df["ineffective_date"] - events_df["effective_date"]
).dt.days

transfer_rows = events_df[events_df["event_type"] == "transfer"].copy()
if transfer_rows.empty:
    print("No transfer rows to evaluate durations.")
else:
    duration_summary = (
        transfer_rows.groupby("first_event_type")["transfer_duration_days"]
        .agg(
            transfers="size",
            zero_duration=lambda s: int((s == 0).sum()),
            zero_share=lambda s: float((s == 0).mean()),
            median_days="median",
            p10_days=lambda s: float(s.quantile(0.1)),
            p90_days=lambda s: float(s.quantile(0.9)),
            negative_count=lambda s: int((s < 0).sum()),
        )
        .sort_index()
    )
    display(duration_summary)
```

```python
# Snapshot row count estimate for 2025 with active residents
base_dfs = dfs_filtered if "dfs_filtered" in globals() else dfs
residents = base_dfs.get("residents", pd.DataFrame()).copy()

if residents.empty:
    raise ValueError("residents table is missing or empty.")

from typing import Optional

def pick_first_col(df: pd.DataFrame, candidates: list[str]) -> Optional[str]:
    for c in candidates:
        if c in df.columns:
            return c
    return None

admit_col = pick_first_col(
    residents,
    ["admission_date", "admit_date", "admitted_at", "admit_at", "start_at"],
 )
discharge_col = pick_first_col(
    residents,
    ["discharge_date", "discharged_at", "discharge_at", "end_at", "end_date"],
 )

if admit_col is None:
    raise ValueError("No admission date column found in residents.")
if discharge_col is None:
    raise ValueError("No discharge date column found in residents.")

residents["admit_ts"] = pd.to_datetime(residents[admit_col], errors="coerce")
residents["discharge_ts"] = pd.to_datetime(residents[discharge_col], errors="coerce")

date_start = pd.Timestamp("2025-01-01")
date_end = pd.Timestamp("2025-12-31")

def estimate_rows(step_days: int) -> dict:
    snapshots = pd.date_range(date_start, date_end, freq=f"{step_days}D")
    admits = residents["admit_ts"]
    discharges = residents["discharge_ts"]
    active_counts = []
    for d in snapshots:
        active = (admits <= d) & (discharges.isna() | (discharges >= d))
        active_counts.append(int(active.sum()))
    return {
        "step_days": step_days,
        "snapshots": int(len(snapshots)),
        "total_rows": int(sum(active_counts)),
        "avg_active_per_snapshot": float(np.mean(active_counts)) if active_counts else 0.0,
        "max_active_per_snapshot": int(max(active_counts)) if active_counts else 0,
    }

row_estimates = pd.DataFrame([estimate_rows(7), estimate_rows(15)])
display(row_estimates)
```
